```python
# All libraries & Data points
import jetset
print('tested on jetset',jetset.__version__)
import warnings
warnings.filterwarnings('ignore')
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from jetset.data_loader import Data
from jetset.data_loader import ObsData
from jetset.plot_sedfit import PlotSED
from jetset.jet_model import Jet
from jetset.minimizer import fit_SED,ModelMinimizer
from jetset.model_manager import  FitModel
from jetset.sed_shaper import  SEDShape
from jetset.obs_constrain import ObsConstrain
from jetset.minimizer import fit_SED
from jetset.template_2Dmodel import EBLAbsorptionTemplate
from jetset.mcmc import McmcSampler

ebl_franceschini=EBLAbsorptionTemplate.from_name('Franceschini_2008')
sed_data=ObsData.load('J163547_data.pkl')
plt.rcParams.update({'font.size': 14})
plt.rcParams['figure.dpi'] = 200
plt.rcParams["savefig.bbox"] = 'tight'
```

```python
model=FitModel.load_model("fit_model_5_3.pkl")

model.add_component(ebl_franceschini)

model.link_par(par_name='z_cosm',
                      from_model='Franceschini_2008',
                      to_model='jet_leptonic')

model.composite_expr='(jet_leptonic)*Franceschini_2008'

model.eval()
#model.save_model('EBL5.pkl')
#model.show_model()
p2=model.plot_model(sed_data=sed_data)
legend_labels = [r'$\eta=0.057$', 'Sync', 'SSC', 'EC Disc', 'Disc', 'EC DT', 'DT','EC BLR', 'EBL Corrections', 'Data']
p2.sedplot.legend(labels = legend_labels, loc='upper left', fontsize=12)
p2.sedplot.grid(False)
p2.setlim(x_min=1e8,y_max=3.0e-11, y_min=7e-16)
p2.save('img_EBL/EBL5.png')
```
